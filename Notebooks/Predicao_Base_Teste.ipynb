```python
import pandas as pd
from sklearn import preprocessing
from pycaret.classification import load_model, predict_model
```

```python
df_teste = pd.read_csv('C:\\Desafio Cientista de Dados - Indicium\\Dados\RAW\\desafio_manutencao_preditiva_teste.csv')
df_teste
```

```python
df_teste.drop(['udi', 'product_id'], axis=1, inplace=True)
```

```python
df_teste
```

### Aplicando Label Encoder na feature "type"

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
df_teste['type']= label_encoder.fit_transform(df_teste['type'])
```

```python
df_teste
```

```python
modelo = load_model('model')
```

```python
df_predicoes = predict_model(modelo, raw_score = False, data = df_teste)
```

```python
df_predicoes.drop(['type', 'air_temperature_k', 'process_temperature_k',
       'rotational_speed_rpm', 'torque_nm', 'tool_wear_min', 'Score'], axis=1, inplace=True)
```

```python
df_predicoes
```

```python
df_predicoes.rename(columns={'Label': 'predictedValues'}, inplace = True)
```

```python
linhas = df_predicoes.index
```

```python
df_predicoes.insert(0,"rowNumber",linhas,True)
```

```python
df_predicoes
```

**Salvando os dados em formato CSV**

```python
df_predicoes.to_csv('predicoes.csv', index=False)
```
